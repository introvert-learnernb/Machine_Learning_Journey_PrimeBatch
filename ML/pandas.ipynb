```python
#panda series are homogeneous (same data type only stores..)
s1 = pd.Series([1,2,3])
s2 = pd.Series([10,20])  #vectorized Operations

print(s1+s2)

s2[2] = 30  #panda series are mutable but size is not mutable..
print(s2)

new_s2 = s2.drop(0)
print(new_s2)

#series handles missing values with nan
```
